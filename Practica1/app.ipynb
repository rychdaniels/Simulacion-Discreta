```python
import matplotlib.pyplot as plt
import numpy as np
import random

arrLanzamiento = [];
arrNumJuegos = [];

arrGanancia = [];

# Clase encargada de incializar todos los datos necesarios para el juego
class Juego:
    
    def __init__(self, ganancia=0, gana=0, pierde=0, tiros=0, tirosT=0):
          
        self.ganancia = ganancia; #Muestra la ganancia, si no hay ganacia se muestra el numero en negativo
        self.tiros = tiros; #Son lo tiros por un solo juego
    
#Metodo para simular el lanzamiento de la moneda     
    def lanzarMoneda(self):        
        dinero = 8; # Dinero que el participantes siempre tiene al final del juego
        caras = 0; # Contador para saber el numero de veces que cayó cara
        cruz = 0; # Contador para saber el numero de veces que cayó cara
        self.tiros = 0; #Cada que iniciamos un nuevo juego, los tiros del juego deben reiniciarse
        diferencia = 0;
        self.ganancia = 0;
        
        
        while abs(diferencia) < 3:
            num = np.random.uniform(0.0, 1.0)
            if num < 0.5:                
                caras += 1               
            else:                
                cruz += 1
            
            #Aumentamos el conteo de los tiros
            
            self.tiros += 1           
            diferencia = caras - cruz  
        
        
        self.ganancia = dinero - self.tiros
        
#         print("Caras ", caras)
#         print("Cruz", cruz)    
#         print("Numero de lanzamientos por juego", self.tiros)        
#         print("La ganancia es de: $" , self.ganancia)    


        
def creaPartida(numJuegos):       
    juego = Juego();
    promedioTiros = 0;
    promedioGanancia = 0;
    for i in range(numJuegos):        
        juego.lanzarMoneda()          
        promedioTiros += juego.tiros/numJuegos
        promedioGanancia += juego.ganancia/numJuegos
    
    arrNumJuegos.append(numJuegos)
    arrGanancia.append(promedioGanancia)
    arrLanzamiento.append(promedioTiros)
  
    
#     print ("Promedio de tiros por partida ", promedioTiros)
#     print ("Tiros totales por partida ", juego.tirosT)
#     print ("Gana ",juego.gana)
#     print ("Pierde ", juego.pierde)

def dibujaGrafica(juegosPartida, promedioPartida):
    plt.xlabel("Número de juegos por partida")
    plt.ylabel("Número promedio de juegos ganados")
    plt.plot(juegosPartida, promedioPartida)
    plt.show()
    
def dibujaGraficaLanzamientos(juegosPartida, promedioPartida):
    plt.xlabel("Número de juegos por partida")
    plt.ylabel("Número promedio lanzamientos")
    plt.plot(juegosPartida, promedioPartida)
    plt.show()
    
def main():    
   
    creaPartida(10)
    creaPartida(30)
    creaPartida(50)
    creaPartida(80)
    creaPartida(100)
    creaPartida(100)
    creaPartida(200)
    creaPartida(300)
    creaPartida(400)
    creaPartida(500)
    creaPartida(600)
    creaPartida(700)
    creaPartida(800)
    creaPartida(900)
    creaPartida(1000)
    creaPartida(1200)
    creaPartida(1300)
    creaPartida(1400)
    creaPartida(1500)
    creaPartida(1600)
    creaPartida(1700)
    creaPartida(1800)
    creaPartida(1900)
    creaPartida(2000)
    creaPartida(3000)
    creaPartida(4000)
    creaPartida(5000)
    creaPartida(9000)
    creaPartida(10000)

 
    dibujaGrafica(arrNumJuegos,arrGanancia)
    dibujaGraficaLanzamientos(arrNumJuegos,arrLanzamiento)
        
main()
```
