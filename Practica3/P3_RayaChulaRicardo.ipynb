```python
######################################################################################
#    Ricardo Raya Chula                                                              #
#    Descripción: Programa que simula la llegada y salida de clientes de un banco,   #
#                 en base al calculo de eventos futuros, en este caso próximas       #
#                 llegadas o próximas salidas.                                       #
######################################################################################


import matplotlib.pyplot as plt
import numpy as np
import random # Genera numeros aleatorios

import math # Para poder ocupar y calcular el ln(logaritmo natural)


#Metodo que calcula la siguiente llegada
def proximaLlegada(dt):
    rA = np.random.uniform(0.0, 1.0) #rA: Numero aleatorio para cuando llega un cliente
    tHoras = (math.log10(1 - rA)) / (-3) #Formula para calcular la proxima entrada en horas
    tMinutos = tHoras * 60 #Convertimos a minutos
    return dt + tMinutos

#Metodo que calcula la siguiente salida
def proximaSalida(dt):
    rD = np.random.uniform(0.0, 1.0) #rD: Numero aleatorio para cuando llega un cliente
    tHoras = (math.log10(1 - rD)) / (-5) #Formula para calcular la proxima salida en horas
    tMinutos = tHoras * 60 #Convertimos a minutos
    return dt + tMinutos

#Metodo para regresar el valor más pequeño
def menor(llegada, salida, dt):
    if(salida > dt and llegada > dt):
        
        if(llegada > salida):        
            return "salida"        
        else:            
            return "llegada"
    
    
    if (llegada > dt) and (salida <= dt):
        return 'llegada'
    if (llegada <= dt) and (salida > dt):   
        return 'salida'


def simulacion(iteraciones):
    nt = 0 #Representa el estado del sistema
    contador_Clientes = 0; #Contador de clientes en el sistema 
    dt= 0 # Será el aumento de tiempo del programa(t)   
    hllegada = proximaLlegada(dt); #El primer evento siempre sera una llegada
    hSalida = 0; #Al principio no habra salidas
    evento = "llegada" #Guarda el siguiente evento
    
    
    print("t                                nt                hlleg                            hsal               Siguiente Evento")
    print(dt,"                      ",nt,"                      ",hllegada,"                    ",hSalida,"         ",evento      )
    dt = hllegada
    
    # El while representa el horario del banco
    while (contador_Clientes < iteraciones or nt > 0):                    
        
        
        # Cuando llega un cliente        
        if(evento == "llegada"):                        
            contador_Clientes += 1 #Aumentamos el contaodor de clientes 
            nt += 1 #Aumetamos el estado del sistema
            dt = hllegada  #Acutalizamos la hora de llegada
            hllegada = proximaLlegada(dt); #Asiganamos la nueva hor de llegada             
            if(contador_Clientes > 0 and hSalida < dt): # Debe haber clientes y estar abierto el banco para generar salidas
                hSalida = proximaSalida(dt)                              
                
            
        #Cuando un cliente se va             
        if (evento == "salida"):
            nt -= 1 # Restamos al estado del sistema
            dt = hSalida #Actualizamos el valor de dt
            if( hllegada < dt and nt == 0): # Sla hora de llegada actual es menor  dt y no hay nadie en el sistema generamos llegadas
                hllegada = proximaLlegada(dt)             
            
            if(nt > 0 ): #Si hay clientes en el sistema, generamos una salida                       
                hSalida = proximaSalida(dt)     
        
        # De acuerdo a los parametros guardamos el evento mas cercano
        evento = menor(hllegada, hSalida, dt)   
        
        
        #Se imprimen los resultados
        print(dt,"              ",nt,"             ",hllegada,"            ",hSalida,"              ",evento)
       
simulacion(10)



```
